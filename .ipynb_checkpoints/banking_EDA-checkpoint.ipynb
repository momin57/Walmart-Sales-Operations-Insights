```python
import pyodbc

# Connect to SQL Server using Windows Authentication
conn = pyodbc.connect(
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=MOMIN;'
    'DATABASE=banking_project;'     # Replace with the name of your actual database 
    'Trusted_Connection=yes;'
)
```

```python
import pandas as pd

query = "SELECT * FROM customer"
df = pd.read_sql(query, conn)

df.head()
```

## Catgorical Analysis

```python
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount_of_Credit_Cards", "Nationality", "Occupation", "Fee_Structure", "Loyalty_Classification", "Properties_Owned", "Risk_Weighting", "Income_Band"]]

for col in categorical_cols:
    print(f"Value counts for '{col}':")
    print("")
    value_counts = df[col].value_counts()
    display(value_counts)
    
    # Plot
    value_counts.plot(kind='bar', title=col)
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
```

```python
for i, predictor in enumerate(df):
    plt.figure(i)
    sns.countplot(data = df, x = predictor, hue = "GenderId")
```

```python
for i, predictor in enumerate(df):
    plt.figure(i)
    sns.countplot(data = df, x = predictor, hue = "Nationality")
```

```python
for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize = (8, 4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

## Numerical Analysis

```python
numerical_cols = ['Estimated_Income', 'Superannuation_Savings', 'Credit_Card_Balance', 'Bank_Loans', 'Bank_Deposits', 'Checking_Accounts', 'Saving_Accounts', 'Foreign_Currency_Account', 'Business_Lending']

plt.figure(figsize = (15, 25))
for i, col in enumerate(numerical_cols):
    plt.subplot(4, 3, i+1)
    sns.histplot(df[col], kde = True)
    plt.title(col)
plt.show()
```

## Heatmaps

```python
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize = (12, 12))
sns.heatmap(correlation_matrix, annot = True, cmap = 'crest', fmt = '.2f')
plt.title("Correlation Matrix")
plt.show()
```

## EDA Insights

1. Strong Relationships

3. Weak or No Relationships
